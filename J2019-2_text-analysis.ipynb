```python
def count_words(entry):
    words = entry.replace("\n\n", " ")
    words = words.split(" ")
    count = 0
    for word in words:
        if "-" in word or ":" in word or "_" in word:
            count += 2
        elif word == "—" or word == "":
            count += 0
        elif word.endswith("."):
            count += 1
        else:
            count += 1
    return count
```

```python
def count_all_words(entry):
    words = entry.replace("\n\n", " ")
    words = words.split(" ")
    count = 0
    for word in words:
        if "-" in word or ":" in word or "_" in word:
            count += 2
            if "-" in word:
                small_count = 0
                for letter in word:
                    if letter == "-":
                        small_count += 1
                        if small_count == 2:
                            count += 1
                        if small_count == 3:
                            count += 2
        elif word == "—" or word == "":
            count += 0
        elif word.endswith("."):
            count += 1
        else:
            count += 1
    return count
```

```python
def count_word(target_word, target_entry):
    count = 0
    for word in target_entry:
        if word == target_word:
            count += 1
    return count
```

```python
def find_category(target_word2):
    if target_word2 in people:
        return "People", target_word2
    if target_word2 in places:
        return "Places", target_word2
    if target_word2.lower() in themes:
        return "Themes", target_word2.lower()
    if target_word2.lower() in feelings:
        return "Feelings", target_word2.lower()
```
